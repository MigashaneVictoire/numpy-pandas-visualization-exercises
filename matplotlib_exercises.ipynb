Use matplotlib to plot the following equation:

```
y = x^2 -x +2
```
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

````python
x = np.arange(-20,20, 1)
y = ((x**2) - x) + 2

plt.figure(figsize = (5,3))
plt.plot(x,y, c="gray", ls="--")

plt.annotate("Point (0,0)", 
             xy=(0,0), 
             c="b", xytext=(5,300), 
             arrowprops= dict(facecolor= "b"))

plt.title("Rand X vs Computed Y")
plt.xlabel("X")
plt.ylabel("Y")

# plt.savefig("figure1.png")
````

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
```
* y = $\sqrt{x}
* y = x^3
* y = 2^x
* y = 1/(x + 1)
```
You can use functions from the math module to help implement some of the equations above.

````python
x = np.arange(1, 100, 5)

# set y variable
y1 = np.sqrt(x)

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y1, c="b", marker = "o")

plt.title('Random x vs $\sqrt{x}$')
plt.xlabel("Random x")
plt.ylabel("$\sqrt{x}$")
plt.xticks([30,60,90])

plt.grid(":")
````

````python
x = np.arange(1, 100, 5)

# set y variable
y2 = x**3

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y2, c="b", marker = "*")

plt.title('Random x vs $x^3$')
plt.xlabel("Random x")
plt.ylabel("$x^3$")
plt.xticks([30,60,90])

plt.grid("--")
````

````python
x = np.arange(1, 20, 1)

# set y variable
y3 = 2**x

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y3, c="b")

plt.title('Random x vs $2^x$')
plt.xlabel("Random x")
plt.ylabel("$2^x$")
plt.xlim(10)

plt.grid("--")
````

````python
x = np.arange(1, 20, 1)

# set y variable
y4 = 1/(x + 1)

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y4, c="b")

plt.title('Random x vs $1/(x + 1)$')
plt.xlabel("Random x")
plt.ylabel("$1/(x + 1)$")
plt.xticks([5,10,15])

plt.grid("--")
````

Combine the figures you created in the last step into one large figure with 4 subplots.

````python
x = np.arange(1,100, 5)

# set plot size
fig, axis = plt.subplots(2, 2 ,figsize=(7,7))

# set y variable
y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1/(x + 1)

my_y = [y1, y2, y3, y4]
# titles = ["Random x vs $\sqrt{x}$", "Random x vs $x^3$", 'Random x vs $2^x$',"Random x vs $1/(x + 1)"]
y_labels = ["$\sqrt{x}$", "$x^3$", "$2^x$","$1/(x + 1)$"]


counter = 0

# create plots
for row in axis:
    for col in row:
        col.plot(x, my_y[counter])
        
        col.set_xlabel("X")
        counter += 1

# Plot on the first subplot
axis[0, 0].set_title("Random x vs square y")
axis[0,0].set_ylabel("sqrt(x)")

# # Plot on the second subplot
axis[0, 1].set_title("Random x vs cube x")
axis[0,1].set_ylabel("x^3")

# # Plot on the third subplot
axis[1, 0].set_title("Random x vs x power")
axis[1,0].set_ylabel("2^x")

# # Plot on the fourth subplot
axis[1, 1].set_title("Random x vs over x + 1")
axis[1,1].set_ylabel("1/(x + 1)")

plt.tight_layout()
````

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

````python
x = np.arange(1,100, 5)

# set plot size
plt.figure(figsize=(6,3))

# plt.plot(x, y1, color = 'gray', ls = '--', marker = '^', label = '$\sqrt{x}$')
plt.plot(x, y1, color = 'gray', ls = '--', label = '$\sqrt{x}$')
plt.plot(x, y2, color = 'blue', ls = '--', label = '$x^3$')
plt.plot(x, y3, color = 'red', ls = '--', label = '$2^x$')
plt.plot(x, y4, color = 'green', ls = '--', label = '$1/(x + 1)$')

plt.title('X vs Changes in y')
plt.xlabel('X')
plt.ylabel('Change in Y')

plt.ylim(-3,20)
plt.xlim(0, 20)
plt.legend()
````
