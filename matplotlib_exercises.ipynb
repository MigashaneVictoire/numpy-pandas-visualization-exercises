You can use functions from the math module to help implement some of the equations above.

````python
x = np.arange(76,200, 4)

# set y variable
y1 = np.sqrt(x)

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y1)

# plt.title('Random x vs square y')
plt.xlabel("Random x")
plt.ylabel("sqrt(x)")
````

````python
x = np.arange(76,200, 4)

# set y variable
y2 = x**3

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y2)

# plt.title('Random x vs cube x')
plt.xlabel("Random x")
plt.ylabel("x^3")
````

````python
x = np.arange(76,200, 4)

# set y variable
y3 = 2**x

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y3)

# plt.title('Random x vs x power')
plt.xlabel("Random x")
plt.ylabel("2^x")
````

````python
x = np.arange(76,200, 4)

# set y variable
y4 = 1/(x + 1)

# Plot on the first subplot
plt.figure(figsize=(5,3))
plt.plot(x,y4)

# plt.title('Random x vs over x + 1')
plt.xlabel("Random x")
plt.ylabel("1/(x + 1)")
````

Combine the figures you created in the last step into one large figure with 4 subplots.

````python
x = np.arange(76,200, 4)

# set plot size
fig, axis = plt.subplots(2, 2 ,figsize=(7,7))

# set y variable
y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1/(x + 1)

my_y = [y1, y2, y3, y4]
counter = 0

# create plots
for row in axis:
    for col in row:
        col.scatter(x, my_y[counter])
        counter += 1

# Plot on the first subplot
axis[0, 0].set_title("Random x vs square y")
axis[0, 0].set_xlabel("Random x")
axis[0,0].set_ylabel("sqrt(x)")

# # Plot on the second subplot
axis[0, 1].set_title("Random x vs cube x")
axis[0, 1].set_xlabel("Random x")
axis[0,1].set_ylabel("x^3")

# # Plot on the third subplot
axis[1, 0].set_title("Random x vs x power")
axis[1, 0].set_xlabel("Random x")
axis[1,0].set_ylabel("2^x")

# # Plot on the fourth subplot
axis[1, 1].set_title("Random x vs over x + 1")
axis[1, 1].set_xlabel("Random x")
axis[1,1].set_ylabel("1/(x + 1)")
````
