# Big O Notation

```python
import matplotlib.pyplot as plt
import math as math
import numpy as np
import pandas as pd
```

Make a new Jupyter notebook named big_o_notation.ipynb

- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

Curves to graph


- O(1) and label the curve "O(1)"
- O(log n) and label curve as "O(log n)"
- O(n) and label the curve "O(n)"
- O(n long n) and label it "O(n log n)"
- O (n^2) and label it "O(n^2)"
- O (2^n) and label it "O(2^n)"
- O (n!) and label it "O(n!)"
- O (n^n) and label it "O(n^n)"


```python
x = np.arange(1,20,1)

y1 = np.ones(len(x))
y2 = np.log(x)
y3 = x
y4 = x * np.log(x)
y5 = x**2
y6 = 2**x
y7= np.array([math.factorial(i) for i in x])
y8 = x**x

plt.plot(x,y1, marker= "o", label= 'y1= $O(1)$')
plt.plot(x,y2, marker= "o", label= 'y2 = $O(log n)$')
plt.plot(x,y3, marker= "o", label= 'y3= $O(n)$')
plt.plot(x,y4, marker= "o", label= 'y4 = $O(n log n)$')
plt.plot(x,y5, marker= "o", label= 'y5 = $O(n^2)$')
plt.plot(x,y6, marker= "o", label= 'y6 = $O(2^n)$')
plt.plot(x,y7, marker= "o", label= 'y7 = $O(n!)$')
plt.plot(x,y8, marker= "o", label= 'y8 = $O(n^n)$')

plt.xlim(1,5)
plt.ylim(-3,50)
plt.legend()

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
```
